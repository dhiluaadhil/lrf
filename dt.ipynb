```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
import joblib

# Load data
df = pd.read_csv("Movie_Interests_DecisionTree.csv")

# Encode the target 'Interest' (e.g., Action, Animation, Drama)
le_interest = LabelEncoder()
df['Interest'] = le_interest.fit_transform(df['Interest'])

# Features (Age, Gender) and Target (Interest)
X = df[['Age', 'Gender']]
y = df['Interest']

# Train Decision Tree
model = DecisionTreeClassifier()
model.fit(X, y)

# Save the model and the encoder
joblib.dump(model, "movie_model.pkl")
joblib.dump(le_interest, "interest_encoder.pkl")

print("Model and encoder saved successfully!")
```
